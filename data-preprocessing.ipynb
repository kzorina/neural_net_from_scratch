```python
import numpy as np
import pandas as pd
from sklearn import preprocessing

np.set_printoptions(threshold=np.nan) #for displaying arrays without truncating
```

Data was loaded from the next source: https://archive.ics.uci.edu/ml/datasets/Car+Evaluation

It has the next structure:  
 -  buying:       v-high, high, med, low  
 -  maint:        v-high, high, med, low  
 -  doors:        2, 3, 4, 5-more  
 -  persons:      2, 4, more  
 -  lug_boot:     small, med, big  
 -  safety:       low, med, high  

```python
col_names = [
    'buying', 
    'maint',
    'doors',
    'persons',
    'lug_boot',
    'safety',
    'evaluation'
]
categorical_cols = [
    'buying', 
    'maint',
    'doors',
    'persons',
    'lug_boot',
    'safety',
    'evaluation'
]
```

```python
data_raw = pd.read_csv("data/car_data/car.data", sep=',', header=None, names = col_names)
le = preprocessing.LabelEncoder()
data_le = data_raw.apply(le.fit_transform)
data = np.asarray(data_le)
data_x = data[:, :-1]
data_y = data[:, -1]
```
